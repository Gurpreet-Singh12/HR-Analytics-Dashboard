```python
# Which department has the highest attrition?
High_dep=df[df['Attrition'] == 'Yes']['Department'].value_counts()
print(High_dep.idxmax(),' department has the highest attrition with',High_dep.max(),'no of Attritions')
```

```python
# Which job role has the highest attrition?
High_job=df[df['Attrition'] == 'Yes']['JobRole'].value_counts()
print(High_job.idxmax(),'job role has the highest attrition with',High_job.max(),'no of Attritions')
```

```python
# Which education field has the highest attrition?
High_edu=df[df['Attrition'] == 'Yes']['EducationField'].value_counts()
print(High_edu.idxmax(),'education field has the highest attrition with',High_edu.max(),'no of Attritions')
```

```python
# Average monthly income and total monthly income by department
monthly_income=df.groupby('Department')['MonthlyIncome'].agg(['sum','mean'])
print(monthly_income)
```

```python
# Average age of employees
avg_age=(df['Age'].sum())/len(df['Age'])
print('The average age of the employees is',avg_age)
```

```python
# Gender distribution and percentage distribution
gender_distribution = df['Gender'].value_counts()
gender_percentage = df['Gender'].value_counts(normalize=True) * 100
print(gender_distribution,'\n',gender_percentage)
```

```python
# Overtime vs Attrition
over_att=df.groupby('OverTime')['Attrition'].value_counts()
print(over_att)
```

```python
# Years at company distribution
years=df.groupby('YearsAtCompany')['YearsAtCompany'].size().sort_index()
print(years)
```

```python
# Job Satisfaction distribution
print(df.groupby('JobSatisfaction').size().sort_values(ascending=False))
```

```python
# Business travel distribution
print(df.groupby('BusinessTravel').size())
```

```python
# Attrition by department
att_dep = df[df['Attrition'] == 'Yes']['Department'].value_counts()
att_dep.plot(kind='bar',color='cyan',edgecolor='green')
plt.title('Attrition by Department',fontdict={'fontsize':20,'fontweight':5,'color':'Red'})
plt.xlabel('Department',fontdict={'fontsize':10,'fontweight':3,'color':'blue'})
plt.ylabel('Attrition',fontdict={'fontsize':10,'fontweight':3,'color':'blue'})
```

```python
# Attrition by Job role
att_role = df[df['Attrition'] == 'Yes']['JobRole'].value_counts()
att_role.plot(kind='bar',color='cyan',edgecolor='green')
plt.title('Attrition by Department',fontdict={'fontsize':20,'fontweight':5,'color':'Red'})
plt.xlabel('JobRoles',fontdict={'fontsize':10,'fontweight':3,'color':'blue'})
plt.ylabel('Attrition',fontdict={'fontsize':10,'fontweight':3,'color':'blue'})
```
